```python
# ---------------------------------
# Import libraries
# ---------------------------------
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder


# ---------------------------------
# Load Diabetes dataset
# ---------------------------------
diabetes = pd.read_csv('Dataset of Diabetes .csv')


# ---------------------------------
# 1. DATA CLEANING
# ---------------------------------

# (a) Handling Missing Values

# Numerical columns -> mean
num_cols = diabetes.select_dtypes(include=['int64', 'float64']).columns
diabetes[num_cols] = diabetes[num_cols].fillna(diabetes[num_cols].mean())

# Categorical columns -> mode
cat_cols = diabetes.select_dtypes(include='object').columns
for col in cat_cols:
    diabetes[col] = diabetes[col].fillna(diabetes[col].mode()[0])


# (b) Handling Categorical Data (Label Encoding)
le = LabelEncoder()
for col in cat_cols:
    diabetes[col] = le.fit_transform(diabetes[col])


# (c) Handling Outliers (IQR Method)
Q1 = diabetes.quantile(0.25)
Q3 = diabetes.quantile(0.75)
IQR = Q3 - Q1

diabetes = diabetes[~((diabetes < (Q1 - 1.5 * IQR)) |
                      (diabetes > (Q3 + 1.5 * IQR))).any(axis=1)]


# ---------------------------------
# 2. DATA TRANSFORMATION
# ---------------------------------

# (a) Min-Max Normalization
minmax = MinMaxScaler()
diabetes_minmax = minmax.fit_transform(diabetes)
diabetes_minmax_df = pd.DataFrame(diabetes_minmax, columns=diabetes.columns)


# (b) Standard Scaling
standard = StandardScaler()
diabetes_std = standard.fit_transform(diabetes)
diabetes_std_df = pd.DataFrame(diabetes_std, columns=diabetes.columns)
```
